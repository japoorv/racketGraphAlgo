```python
import matplotlib.pyplot as plt
import pickle
file=open('./results/time','rb')
time=pickle.load(file)
file=open('./results/memory','rb')
memory=pickle.load(file)
time1=list(map(lambda x: x.split(':'),time))
time2=list(map(lambda x:float(x[0])*60+float(x[1]),time1))
memory=list(map(lambda x : int(x),memory))
```

```python
plt.figure(num=None, figsize=(20, 20), dpi=80, facecolor='w', edgecolor='k')
plt.title('Memory (in Bytes) vs Nodes')
plt.xlabel('No. of Nodes')
plt.ylabel('Memory (in Bytes)')
plt.plot(list(range(1,1000002,10000)),memory,'x')
```

```python
plt.figure(num=None, figsize=(20, 20), dpi=80, facecolor='w', edgecolor='k')
plt.title('Time (sec) vs Nodes')
plt.xlabel('Number of Nodes')
plt.ylabel('Time (seconds)')
plt.plot(list(range(1,1000002,10000)),time2,'o')
```

```python
import pandas as pd
df=pd.DataFrame(time2)
df['Number of Nodes']=list(range(1,1000002,10000))
cols=df.columns.tolist()
cols.append(cols[0])
cols=cols[1:]
df=df[cols]
df.columns=['Number of Nodes','Cumulative Time (seconds)']
df.to_csv('cumulativeTimeVSnodes.csv',index=False,header=True)
```

```python
df=pd.DataFrame(memory)
df['Number of Nodes']=list(range(1,1000002,10000))
cols=df.columns.tolist()
cols.append(cols[0])
cols=cols[1:]
df=df[cols]
df.columns=['Number of Nodes','Max RSS Memory (bytes)']
df.to_csv('maxRSSmemoryVSnodes.csv',index=False,header=True)
```
